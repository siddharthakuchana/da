```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
df=pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

y=diabetes.target
x=range(len(y))

plt.plot(x,y,color="blue")
plt.title("patient disease distribution")
plt.xlabel("patients index")
plt.ylabel("disease index")
plt.show()

plt.scatter(df['bmi'],y,color="skyblue")
plt.title("bmi vs disease distribution")
plt.xlabel("bmi index")
plt.ylabel("disease index")
plt.show()

plt.bar(df.columns,df.mean(),color="maroon")
plt.title("fetaure means")
plt.xticks(rotation=45)

plt.show()
```
